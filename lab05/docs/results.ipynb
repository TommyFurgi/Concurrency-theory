```python
import pandas as pd

df = pd.read_csv('results.csv', header=None, names=["Iterations", "Threads", "Tasks", "AvgTime", "DevTime"])

df['AvgTime'] = (df['AvgTime'] / 1_000_000).round(2)  
df['DevTime'] = (df['DevTime'] / 1_000_000).round(2)

df_sorted = df.sort_values(by='AvgTime')
df_sorted = df_sorted.rename(columns={"AvgTime": "AvgTime (ms)"})
df_sorted = df_sorted.rename(columns={"DevTime": "DevTime (ms)"})

display(df_sorted)
```
