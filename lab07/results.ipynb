```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'results.csv'
df = pd.read_csv(file_path, header=None, names=['N', 'time', 'mode'])

grouped_by_mode = df.groupby('mode')

average_waiting_time = df.groupby(['N', 'mode'])['time'].mean().reset_index()

waiter_data = average_waiting_time[average_waiting_time['mode'] == 'waiter']
asymmetric_data = average_waiting_time[average_waiting_time['mode'] == 'asymmetric']

fig, ax = plt.subplots(figsize=(6, 4))

ax.bar(waiter_data['N'] - 0.2, waiter_data['time'], width=0.4, label='Waiter', color='lightblue', align='center')
ax.bar(asymmetric_data['N'] + 0.2, asymmetric_data['time'], width=0.4, label='Asymmetric', color='orange', align='center')

ax.set_xlabel('N (Liczba filozofów)', fontsize=12)
ax.set_ylabel('Średni czas oczekiwania (ms)', fontsize=12)
ax.set_title('Średni czas oczekiwania dla trybów: Waiter i Asymmetric', fontsize=14)
ax.legend()

plt.xticks(df['N'].unique())  
plt.tight_layout()
plt.show()
```
